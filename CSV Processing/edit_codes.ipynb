```python
import pandas as pd
import urllib.request
```

```python
codes = pd.read_csv("https://raw.githubusercontent.com/christinegu27/State-Sentencing-Project/main/CSV%20Processing/detail_codes.csv")
```

```python
codes=codes.replace({'circuitGender':'Gender','circuitRace':'Race','criminalCaseType':'Case Type','criminalChargeClass':'Charge Class','criminalConcludedBy':'Concluded By','criminalProbationType':'Probation Type'})
```

```python
codes = codes[(codes['Code Type'] == "Gender") | (codes['Code Type'] == "Race") | (codes['Code Type'] == "Case Type") | (codes['Code Type'] == "Charge Class")| (codes['Code Type'] == "Concluded By")| (codes['Code Type'] == "Probation Type")]
```
